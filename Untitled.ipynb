```python
df['quality'].value_counts().nlargest(10).plot(kind='bar', figsize=(10,5))
plt.title('Number of Quality Standard Score')
plt.ylabel('Number of Quality labels/Index')
plt.xlabel('Labels');
```

### Plotting heat map to evaluate the feature dependency 

```python
import seaborn as sns
plt.figure(figsize=(20,10))
cormat = df.corr()
sns.heatmap(cormat,cmap = 'BrBG',annot=True)
cormat
```

### Scatterplot to check the pairwise trend/correlation among the features

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['fixed acidity'],df['density'])
ax.set_xlabel('fixed acidity')
ax.set_ylabel('density')
plt.show()
```

```python
sns.scatterplot(data=df,x = 'fixed acidity',y = 'density')
```
